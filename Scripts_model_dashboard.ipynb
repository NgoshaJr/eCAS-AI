```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np
import pandas as pd
import tempfile

# ========================================
# Load YOLOv8 model
# ========================================
model = YOLO("/content/runs/detect/train/weights/best.pt")  # update path if needed

# ========================================
# Pixel-to-length conversion scales
# ========================================
# You can adjust the scale and units as per your calibration
scales = {
    "L.miodon": {"scale": 11.35, "unit": "mm"},
    "L.stapesii": {"scale": 78.08, "unit": "cm"},
    "S.tanganicae": {"scale": 11.00, "unit": "mm"}
}


# ========================================
# Detection and Measurement Function
# ========================================
def detect_and_measure(input_image):
    # Convert PIL Image to OpenCV format
    img = cv2.cvtColor(np.array(input_image), cv2.COLOR_RGB2BGR)

    # Run YOLO prediction
    results = model.predict(source=img, conf=0.5, verbose=False)[0]

    results_data = []

    # Process each detected object
    for box in results.boxes:
        cls_id = int(box.cls[0])
        cls_name = model.names[cls_id]
        conf = float(box.conf[0])

        # Only process species defined in 'scales'
        if cls_name not in scales:
            continue

        # Get bounding box dimensions
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        length_px = max(x2 - x1, y2 - y1)

        # Convert pixels to real length
        scale = scales[cls_name]["scale"]
        unit = scales[cls_name]["unit"]
        length = round(length_px / scale, 2)

        # Save detection result
        results_data.append({
            "Species": cls_name,
            "Length": length,
            "Unit": unit,
            "Confidence": round(conf, 2)
        })

        # Draw bounding box & label on image
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        label = f"{cls_name}: {length} {unit} ({conf:.2f})"
        cv2.putText(img, label, (x1, max(y1 - 10, 20)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Convert annotated image back to RGB for Gradio
    annotated_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Create results DataFrame
    if results_data:
        df = pd.DataFrame(results_data)
    else:
        df = pd.DataFrame(columns=["Species", "Length", "Unit", "Confidence"])

    # Save CSV temporarily for download
    tmp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".csv")
    df.to_csv(tmp_file.name, index=False)

    return annotated_img, df, tmp_file.name


# ========================================
# Gradio Interface
# ========================================
iface = gr.Interface(
    fn=detect_and_measure,
    inputs=gr.Image(type="pil"),
    outputs=[
        gr.Image(type="numpy", label="Annotated Image"),
        gr.DataFrame(headers=["Species", "Length", "Unit", "Confidence"], label="Detection Results"),
        gr.File(label="Download CSV Results")
    ],
    title="eCAS Fish Detector",
    description="Uploaded images are resized to 640x640 for detection. Bounding boxes are displayed in original size, lengths remain calibrated.",
    allow_flagging="never"
)

iface.launch(share=True)
```

```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np
import pandas as pd
import tempfile

# ================================
# Load YOLO model
# ================================
model = YOLO("/content/runs/detect/train/weights/best.pt")  # Trained on 640x640 images

# ================================
# Pixel-to-length scales (calibrated for 640x640)
# ================================
TRAINING_SIZE = (4080, 3060)  # (width, height)
scales = {
    "L.miodon": {"scale": 11.35, "unit": "mm"},
    "L.stapesii": {"scale": 78.08, "unit": "cm"},
    "S.tanganicae": {"scale": 11.00, "unit": "mm"}
}


def detect_and_measure(input_image):
    """
    Detect fish, estimate lengths based on 4080x3060 calibration,
    and display bounding boxes scaled to original uploaded image size.
    """
    # Convert PIL â†’ OpenCV
    img_orig = cv2.cvtColor(np.array(input_image), cv2.COLOR_RGB2BGR)
    orig_h, orig_w = img_orig.shape[:2]
    train_w, train_h = TRAINING_SIZE

    # Compute scaling factors for width and height
    scale_w = orig_w / train_w
    scale_h = orig_h / train_h

    # Resize image to 4080x3060 for YOLO detection
    img_resized = cv2.resize(img_orig, TRAINING_SIZE)

    # Run YOLO prediction
    results = model.predict(source=img_resized, conf=0.5, imgsz=640, verbose=False)[0]

    detections = []

    for box in results.boxes:
        cls_id = int(box.cls[0])
        cls_name = model.names[cls_id]
        conf = float(box.conf[0])

        if cls_name not in scales:
            continue

        # Bounding box coordinates in 4080x3060 image
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        length_px = max(x2 - x1, y2 - y1)

        # Length estimation using calibration
        scale = scales[cls_name]["scale"]
        unit = scales[cls_name]["unit"]
        length = round(length_px / scale, 2)

        # Save detection data (clean output)
        detections.append({
            "Species": cls_name,
            "Length": length,
            "Unit": unit,
            "Confidence": round(conf, 2)
        })

        # Scale bounding boxes back to original uploaded image size
        x1_orig = int(x1 * scale_w)
        x2_orig = int(x2 * scale_w)
        y1_orig = int(y1 * scale_h)
        y2_orig = int(y2 * scale_h)

        # Draw bounding box on original image
        cv2.rectangle(img_orig, (x1_orig, y1_orig), (x2_orig, y2_orig), (0, 255, 0), 3)

        # Draw label
        label = f"{cls_name}: {length} {unit} ({conf:.2f})"
        (text_w, text_h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 1, 2)
        cv2.rectangle(img_orig, (x1_orig, y1_orig - text_h - 8),
                      (x1_orig + text_w, y1_orig), (0, 255, 0), -1)
        cv2.putText(img_orig, label, (x1_orig, y1_orig - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Convert annotated image back to RGB for Gradio
    annotated_img = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)

    # Prepare DataFrame
    if detections:
        df = pd.DataFrame(detections)
    else:
        df = pd.DataFrame(columns=["Species", "Length", "Unit", "Confidence"])

    # Save CSV temporarily
    tmp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".csv")
    df.to_csv(tmp_file.name, index=False)

    return annotated_img, df, tmp_file.name


# ================================
# Gradio Interface
# ================================
iface = gr.Interface(
    fn=detect_and_measure,
    inputs=gr.Image(type="pil"),
    outputs=[
        gr.Image(type="numpy", label="Annotated Image"),
        gr.DataFrame(headers=["Species", "Length", "Unit", "Confidence"], label="Detection Results"),
        gr.File(label="Download CSV Results")
    ],
    title="eCAS-AI Fish Detector (Accurate Scaling)",
    description="Uploads are resized to 4080x3060 for calibrated detection. Bounding boxes are displayed in original scale with estimated lengths.",
    allow_flagging="never"
)

iface.launch(share=True)
```
